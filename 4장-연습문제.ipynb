```
from google.colab.patches import cv2_imshow
import numpy as np
import cv2
import matplotlib.pyplot as plt
from google.colab import drive

drive.mount('/content/gdrive') 
#영상 불러오기
img = cv2.imread('/content/gdrive/My Drive/Image_Processing/girl.jpg')

YCrCb_img = cv2.cvtColor(img, cv2.COLOR_BGR2YCrCb) 
YCrCb_planes = cv2.split(YCrCb_img)

#히스토그램 평활화 (output_image를 따로 정의하지 말고 동적으로 생성하자)
YCrCb_planes[0] = cv2.equalizeHist(YCrCb_planes[0])

YCrCb_out = cv2.merge(YCrCb_planes) 
output_img = cv2.cvtColor(YCrCb_out, cv2.COLOR_YCrCb2RGB)

hist=cv2.calcHist([img[0]],[0],None,[256],[0,256])
plt.title("Histogram")
plt.plot(hist,color='b',label='original histo')
plt.xlim([0,256])
hist=cv2.calcHist([YCrCb_planes[0]],[0],None,[256],[0,256])
plt.plot(hist,color='g',label='Equalized histo')
plt.xlim([0,256])
plt.legend()

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) 
#그림을 화면에 출력
plt.figure(figsize=(20,20))# 영상의 크기를 키워주자
plt.subplot(1,2,1)
plt.title("Image1")
plt.imshow(img)
plt.axis("off")
plt.subplot(1,2,2)
plt.title("Equalized Image")
plt.imshow(output_img)
plt.axis("off")
plt.show()
```
