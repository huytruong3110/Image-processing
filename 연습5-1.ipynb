```
from google.colab.patches import cv2_imshow
import numpy as np
import cv2
import matplotlib.pyplot as plt
from google.colab import drive

drive.mount('/content/gdrive') 
#영상 불러오기
img = cv2.imread('/content/gdrive/My Drive/Image_Processing/fce(salt&pepper noise).bmp')

# color영상을 gray영상으로 만들기
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
# 계수가 1로 구성된 3x3커널 만들기
kernel = np.ones((3,3),np.float32)/9
output_img = cv2.filter2D(gray_img,-1,kernel) 

#그림을 화면에 출력
plt.figure(figsize=(20,20))# 영상의 크기를 키워주자
plt.subplot(1,2,1)
plt.title("Image1")
plt.imshow(gray_img, cmap='gray')
plt.axis("off")
plt.subplot(1,2,2)
plt.title("Mean filtering Image")
plt.imshow(output_img, cmap='gray')
plt.axis("off")
plt.show()
```
